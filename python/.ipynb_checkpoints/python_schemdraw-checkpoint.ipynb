### For Loop

A for loop is a fundamental control structure in programming that allows you to repeatedly execute a block of code a specified number of times or over elements in a collection (like a list, string, tuple, dictionary, etc.).

## Let's use IF statement with schemdraw to create Decision Tree

```python
# Libraries
from schemdraw import Drawing
from schemdraw.flow import Start, Arrow, Box, Ellipse, Decision

# Create objects
start = Ellipse().label('Start') 

box_out = Box().label('Go \noutdoors')
box_relax = Box().label('Stay \nhome and relax')
box_work = Box().label('Work \non projects')

box_climbing = Box().label('Go for \nRock Climbing')
box_workout = Box().label('Go for \nWorkout')
box_runing = Box().label('Go for \nRuning')

box_runing_X_Distance = Box().label('X Distance')
box_runing_X_Duration = Box().label('X Duration')

decision = Decision(S='No', E='Yes', W='No').label('Is the \nweather good?')
decision2 = Decision(S='No Fun', E='Fun', W='Fun').label('What to do?')
decision3 = Decision(S='', E='Best', W='Good').label('For how \nLong?')


end = Ellipse().label('Day End')
 
```

```python
def initial_situation(d2):
    d2 += start
    d2 += Arrow().down(1).at(start.S)
    d2 += decision
    d2 += Arrow().down(1).at(decision.S)
    d2 += box_relax
    d2 += Arrow().left(1).at(decision.W)
    d2 += box_work
    d2 += Arrow().right(2).at(decision.E) 
    d2 += box_out
    return d2

```

```python
d2 = Drawing()

d2 = initial_situation(d2)
d2

```

```python
def weather_decision(d2, choose):
    if choose == 'Go_out':
        d2 += Arrow().right(2).at(decision.E)  # Changed to E (East) anchor
        d2 += box_out.color('green')
        d2 += Arrow().down(6).at(box_out.S)
        d2 += decision2
        d2 += Arrow().right(1).at(decision2.E)  # Changed to E (East) anchor
        d2 += box_climbing
        d2 += Arrow().down(1.5).at(decision2.S)
        d2 += box_workout
        d2 += Arrow().left(3).at(decision2.W)
        d2 += box_runing  #.color('green')
    elif choose == 'relax':
        d2 += Arrow().down(6).at(box_relax.S)
        d2 += end.color('green')

    return d2

draw = weather_decision(d2, choose='Go_out')
draw
```

```python

def what_to_do(d2, choose):
    if choose == 'Runing':
        d2 += Arrow().left(3).at(decision2.W)
        d2 += box_runing.color('green')
        d2 += Arrow().down(5).at(box_runing.S)
        d2 += decision3        
        
        d2 += Arrow().right(1).at(decision2.E)  # Changed to E (East) anchor
        d2 += box_runing_X_Distance
        
        d2 += Arrow().left(1).at(decision3.W)
        d2 += box_runing_X_Duration

    return d2

draw = what_to_do(draw, choose = 'Runing')
draw
```
